```python
import numpy as np
import matplotlib.pyplot as plt
from river import drift
from sklearn.datasets import make_classification

# Function to simulate a 2D classification data stream with drift
def generate_data():
    X, y = make_classification(n_samples=3000, n_features=2, n_redundant=0, n_clusters_per_class=1, flip_y=0, random_state=42)
    # Introducing drift by modifying the data
    X[1000:2000, :] = X[1000:2000, :] * np.array([0.5, 1.5]) + np.array([0.1, -0.1])
    X[2000:, :] = X[2000:, :] * np.array([1.5, 0.5]) - np.array([0.1, 0.1])
    return X, y

X, y = generate_data()

# Create ADWIN drift detectors for each feature
drift_detectors = [drift.ADWIN() for _ in range(X.shape[1])]
drifts = [[] for _ in range(X.shape[1])]

# Process the stream
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        drift_detectors[j].update(X[i, j])
        if drift_detectors[j].drift_detected:
            print(f'Drift detected in feature {j+1} at index {i}')
            drifts[j].append(i)

# Plotting function for the data
def plot_data(X, drifts):
    plt.figure(figsize=(12, 6))
    for i in range(X.shape[1]):
        plt.subplot(1, X.shape[1], i+1)
        plt.plot(X[:, i], label=f'Feature {i+1}')
        if drifts[i]:
            for d in drifts[i]:
                plt.axvline(d, color='red')
        plt.legend()
    plt.show()

plot_data(X, drifts)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from river import drift

# Function to generate the 2D dataset
def generate_data(num_instances, drift_points, drift_width, seed=42):
    np.random.seed(seed)
    features, labels = [], []
    
    for i in range(num_instances):
        # Determine the current distribution based on the drift points
        if i < drift_points[0] or (i >= drift_points[1] + drift_width and i < drift_points[2]):
            mean = [0.8, 0.8]
        elif (i >= drift_points[0] and i < drift_points[0] + drift_width) or (i >= drift_points[2] and i < drift_points[2] + drift_width):
            mean = [0.6, 0.6]
        else:
            mean = [0.4, 0.4]

        x1, x2 = np.random.normal(mean[0], 0.05), np.random.normal(mean[1], 0.05)
        label = 1 if x1 + x2 > 1 else 0
        features.append([x1, x2])
        labels.append(label)
    
    return np.array(features), np.array(labels)

# Generate the dataset
features, labels = generate_data(40000, [10000, 20000, 30000], 1000)

# Drift detection
drift_detector = drift.ADWIN()
drifts = []

for i, (feature, label) in enumerate(zip(features, labels)):
    drift_detector.update(label)   # Here we are using the label for drift detection
    if drift_detector.drift_detected:
        print(f'Change detected at index {i}')
        drifts.append(i)
#         drift_detector.reset()   # Resetting the detector after a drift is detected

# Auxiliary function to plot the detected drifts
def plot_drifts(features, labels, drifts):
    plt.figure(figsize=(10, 6))
    plt.scatter(features[:, 0], features[:, 1], c=labels, alpha=0.5, cmap='coolwarm')
    for drift_detected in drifts:
        plt.axvline(x=drift_detected, color='red', linestyle='--')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('2D Classification Dataset with Detected Drifts')
    plt.show()

plot_drifts(features, labels, drifts)
```
