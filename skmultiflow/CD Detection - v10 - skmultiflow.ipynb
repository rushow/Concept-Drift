```python
import pandas as pd
from skmultiflow.data import DataStream
from skmultiflow.bayes import NaiveBayes
from skmultiflow.trees import HoeffdingTreeClassifier
from skmultiflow.drift_detection import ADWIN, KSWIN, DDM, EDDM, PageHinkley
from skmultiflow.evaluation import EvaluatePrequential
from sklearn.model_selection import train_test_split

def get_data():
    df = pd.read_csv("dataverse/rt_8873985678962563_abrupto.csv")
    return df[['X1', 'X2']], df['class']

# Load the dataset
X, y = get_data()

# Convert the data into a skmultiflow DataStream
stream = DataStream(pd.concat([X, y], axis=1))

# Initialize drift detectors
drift_detectors = {
    'ADWIN': ADWIN(),
    'KSWIN': KSWIN(),
    'DDM': DDM(),
    'EDDM': EDDM(),
    'PageHinkley': PageHinkley()
}

# Initialize base learners
base_learners = {
    'Naive Bayes': NaiveBayes(),
    'Hoeffding Tree': HoeffdingTreeClassifier()
}

# Unfortunately, skmultiflow does not directly use a dict of metrics like river does, so you will need to use the EvaluatePrequential method for evaluation which tracks a set of default metrics.

# Define a function to evaluate models with drift detection
def evaluate_model_with_drift_detection(model, stream, drift_detector):
    # The EvaluatePrequential class in skmultiflow is used to evaluate models. It does not directly accept drift detectors in its API, 
    # so you may need to manually incorporate drift detection logic into model training/testing if you're evaluating drift handling.
    # However, basic usage is as follows, assuming a single model without explicit drift detection in the evaluation loop:
    evaluator = EvaluatePrequential(show_plot=True,
                                    pretrain_size=200,
                                    max_samples=1000,
                                    metrics=['accuracy', 'kappa'])
    evaluator.evaluate(stream=stream, model=model, model_names=['Model'])

# Since skmultiflow's evaluation process is different and doesn't easily allow for manual drift detection within the evaluation loop,
# the detailed implementation of drift handling within the training process as done in the river example is not straightforward.
# The evaluation with skmultiflow is typically more automatic and doesn't involve manual checking and resetting of models based on drift detection.

# As a general example, here's how to evaluate a single model without manual drift detection:
model = base_learners['Naive Bayes'] # Or any other model you've initialized
evaluate_model_with_drift_detection(model, stream, None) # Drift detection logic would need to be integrated separately

# Note: This code shows how to setup a basic evaluation in skmultiflow, but integrating manual drift detection into the training and testing loop as in the river example requires a more customized approach, possibly extending the model classes themselves or handling drift detection outside of the skmultiflow's EvaluatePrequential utility.
```
